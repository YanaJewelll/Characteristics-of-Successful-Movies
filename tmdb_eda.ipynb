```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# saving dataframe
tmdb_df = pd.read_csv('Data/tmdb_data_2001.csv')
tmdb_df.head()
```

```python
# valid financial info count
valid_financial_info_count = ((tmdb_df['budget'] > 0) | (tmdb_df['revenue'] > 0)).sum()
valid_financial_info_count
```

```python
# Question 1: Count movies with valid financial information
valid_financial_info_count = ((tmdb_df['budget'] > 0) | (tmdb_df['revenue'] > 0)).sum()
plt.figure(figsize=(6, 4))
plt.text(0.5, 0.5, f"Movies with valid financial info: {valid_financial_info_count}", fontsize=14, ha='center')
plt.axis('off')
plt.title('Count of Movies with Valid Financial Information')
plt.show()
```

```python
# Question 2: Count movies in each certification category
plt.figure(figsize=(10, 6))
certification_counts = tmdb_df['certification'].value_counts()
certification_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Movies by Certification')
plt.xlabel('Certification')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 3: Average revenue per certification category
plt.figure(figsize=(10, 6))
average_revenue_per_certification = tmdb_df.groupby('certification')['revenue'].mean()
average_revenue_per_certification.plot(kind='bar', color='lightgreen')
plt.title('Average revenue per certification category')
plt.xlabel('Certification')
plt.ylabel('Average Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 4: Average budget per certification category
plt.figure(figsize=(10, 6))
average_budget_per_certification = tmdb_df.groupby('certification')['budget'].mean()
average_budget_per_certification.plot(kind='bar', color='salmon')
plt.title('Average budget per certification category')
plt.xlabel('Certification')
plt.ylabel('Average Budget')
plt.xticks(rotation=45)
plt.show()
```
